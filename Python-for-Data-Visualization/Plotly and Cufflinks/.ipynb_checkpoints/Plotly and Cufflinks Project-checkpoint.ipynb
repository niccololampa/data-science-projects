## Installation

In order for this all to work, you'll need to install plotly and cufflinks to call plots directly off of a pandas dataframe. These libraries are not currently available through **conda** but are available through **pip**. Install the libraries at your command line/terminal using:

    pip install plotly
    pip install cufflinks

** NOTE: Make sure you only have one installation of Python on your computer when you do this, otherwise the installation may not work. **

### Fake Data

```python
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2.head()
```

## Using Cufflinks and iplot()

* scatter
* bar
* box
* spread
* ratio
* heatmap
* surface
* histogram
* bubble

```python
#plotly creates an interactive version of graphs in matplotlib.
#for example matplotlib

df.plot()

```

```python
#then the iplot version which is more interactive

df.iplot()
```

## Scatter

```python
#by default plotly is going to make a line so use mode ='markers'
df.iplot(kind='scatter',x='A',y='B',size=10)
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=10)
```

## Bar Plots

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
#bar plots is for category and values. Thus create an aggregate or groupby on your data to transform it into a form that makes sense to use bar plot. 
#for example df.iplot(kind='bar')

df.count().iplot(kind='bar')
```

## Boxplots

```python
#you can switch off the categories by clicking on the legend. For example b was turned off. 

df.iplot(kind='box')
```

## 3d Surface

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})


df3.iplot(kind='surface',colorscale='rdylbu')

#check plotly documentation for colorscale doesn't exactly match with matplotly colors. 
```

## Spread

```python
#best for financial analysis. 

df[['A','B']].iplot(kind='spread')
```

## histogram

```python
df['A'].iplot(kind='hist',bins=25)
```

```python
#if you run a histogram of all columns you'll get a overlapping histogram and you can turn them on and off to compare each. 

df.iplot(kind='hist',bins=25)
```

```python
#similar to a scatter plot but will change size based on a variable. 
#good for world GDP vs population, happiness factor. UN reports. 

df.iplot(kind='bubble',x='A',y='B',size='C')
```

## scatter_matrix()

Similar to sns.pairplot()

```python
#similar to seaborn's pairplot. 
#make sure all columns are numerical for it to work. 

df.scatter_matrix()
```
