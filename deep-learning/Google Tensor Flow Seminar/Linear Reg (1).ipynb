```python
import tensorflow as tf

slope = tf.Variable(0.0)
intercept = tf.Variable(0.0)
bedroomsize =tf.placeholder(tf.float32)

#linear model

predicted_value = intercept + slope*bedroomsize

actual_value = tf.placeholder(tf.float32)

#loss

loss = tf.reduce_mean(tf.square(predicted_value - actual_value)) #since it has lot of values we reduce to mean. 
```

to minimize loss this is where gradient descent. To figure out where it is to this point U graph. 
learning rate is gaano kalaki step ko from one point to another in U. to get the minimum value. 
 then iterations how many,
 mix learning rate/step and iterations . dapat enough step and iteration to reach the minimum. 

```python
#optimizer - this will change values slope and intercept.

optimizer = tf.train.GradientDescentOptimizer(0.001) # inside is the learning rate/ step


train= optimizer.minimize(loss) #create a tensor. 
```

```python
#run the computational graph
#create a session object

sess = tf.Session()
init= tf.global_variables_initializer()
sess.run(init)
```

```python
#build training loop. 

for i in range(1000): # expirementation here this is the iterations. 
    data = {bedroomsize: room, actual_value:cost}# load the data
    sess.run(train, feed_dict = data)
    
    # to know what is happening
    
    c_slope, c_intercept, c_loss = sess.run([slope,intercept, loss],feed_dict=data)
    
    print("Slope: %s, intercept: %s, cost:%s"%(c_slope, c_intercept, c_loss))
    
```
