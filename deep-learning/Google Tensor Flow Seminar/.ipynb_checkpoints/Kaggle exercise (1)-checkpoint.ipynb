```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset= pd.read_csv('kc_house_data.csv')
```

```python
dataset.head()
```

```python
price= dataset['price']
```

```python
price.head()
```

```python
x= dataset['sqft_living']
```

```python
print(x)
```

```python
plt.plot(price,x,'ro')
```

```python
import tensorflow as tf

slope = tf.Variable(0.0)
intercept = tf.Variable(0.0)
sqliv =tf.placeholder(tf.float32)

#linear model

predicted_value = intercept + slope*sqliv

actual_value = tf.placeholder(tf.float32)

#loss

loss = tf.reduce_mean(tf.square(predicted_value - actual_value))
```

```python

optimizer = tf.train.GradientDescentOptimizer(0.0000001) # inside is the learning rate/ step


train= optimizer.minimize(loss) #create a tensor. 
```

```python
sess = tf.Session()
init= tf.global_variables_initializer()
sess.run(init)
```

```python
for i in range(10000): # expirementation here this is the iterations. 
    data = {sqliv: x, actual_value: price}# load the data
    sess.run(train, feed_dict = data)
    
    # to know what is happening
    
    c_slope, c_intercept, c_loss = sess.run([slope,intercept, loss],feed_dict=data)
    print("Slope: %s, intercept: %s, cost:%s"%(c_slope, c_intercept, c_loss))
```
